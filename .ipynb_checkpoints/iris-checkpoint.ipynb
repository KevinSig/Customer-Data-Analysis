# KNN Model

Now that our data is prepped, we can apply a classification model to try to classify unknown input samples.

In this case, we have four input attributes, and three potential classes. To start, we'll use a **KNN Model**, which stands for K Nearest Neighbors. KNN falls in the category of **supervised learning** algorithms, which means that it takes a labelled set of training observations (consisting of ($X$, $y$)), and attempts to capture the relationship between $X$ and $y$. Given an unseen observation $\hat{x}$, our learned relationship h(x) can hopefully predict the corresponding output $\hat{y}$.

KNN is **instance-based**:
 - Instance-based: Our algorithm doesn't explicitly learn a model. Instead, it memorizes the training instances and uses exactly those instances as the foundation of its knowledge for the prediction process.
 
Since no model is constructed and trained, you cannot abstract away your input data. This means that using a KNN comes at a memory and computational cost compared to parametric models. Classifying any given observation requires recalling the whole input dataset and running through calculations for it. However, it's a lot easier to implement and understand.

# So how does it work?


For classification, K-nearest neighbors essentially is just forming a majority vote between the K most similar instances to a given "unseen" observation. Similarity is defined by some distance metric between two data points. Typically this is Euclidean distance, given by

$d(x,x') = \sqrt{(x_1-x'_1)^2 + (x_2 - x'_2)^2  +  ...  +  (x_n - x'_n)^2}$

More formally, the steps that the algorithm walks through to classify are as follows:
1. Run through the whole dataset computing distance $d$ between unseen observation $\hat{x}$ and training dataset $X$.
2. Identify the K point in the training data that are closest to $\hat{x}$
3. Evaluate the conditional probability for each available class. That is, the fraction of points in set K with each given class label.
4. Finally, input $\hat{x}$ gets assigned to the class with the largest probability.

In simple terms,
> KNN searches the memorized training observations for the K instances that most closely resemble the new instance and assigns to it their most common class

The last question you're probably wondering is: how do we pick K? This is the only hyperparameter that needs tuning for this algorithm, and you can find the ideal value through testing. In general, small K provides the most flexible "fit" of a decision boundary to your data - this provides low bias but high variance. Graphically, that would look like a jagged decision boundary. 




On the other hand, a higher K value averages more voters in the prediction process and is therefore more resilient to outliers. 
This means smoother decision boundaries - lower variance but increased bias.

Okay now we write the predict method which must: 
1. Compute euclidean distance between any new observation and all points in the training set
2. Select K nearest points
3. Take majority vote of classes from K
4. Assign corresponding class to the new observation



```python
from collections import Counter 

def predict(X_train, y_train, x_test, k):
    distances = []
    labels = []
    
    for i in range(len(X_train)):
        distance = np.sqrt(np.sum(np.square(x_test - X_train.iloc[i])))
        distances.append([distance, i])
    
    distances = sorted(distances)
    
    for i in range(k):
        index = distances[i][1]
        labels.append(y_train.iloc[index])
        
    return Counter(labels).most_common(1)[0][0]
```

```python
def KNearestNeighbor(X_train, y_train, X_test, k):
    train(X_train, y_train)
    
    predictions = []
    
    for i in range(len(X_test)):
        predictions.append(predict(X_train, y_train, X_test.iloc[i], k))
    return predictions
    
```
